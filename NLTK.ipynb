```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
set(stopwords.words('english'))
```

```python
data_new = pd.DataFrame({'description':m})
data_new.description.values.tolist()
```

```python
content = data_new.description.values.tolist()
content[2]
```

```python
stop_words = set(stopwords.words('english'))
filtered_sentence = []

for line in content:
    pattern = re.compile(r'\r|\n|\(|\)|\?|\=|\:|\-|\"|\#|\,|\/|\||\d|\.')
    new = re.sub(pattern,'',line)
    word_tokens = word_tokenize(new)
    temporary = []
    for w in word_tokens:
        if w not in stop_words:
            temporary.append(w)
    filtered_sentence.append(temporary)
```

```python
" ".join(filtered_sentence[0])
```

```python
df = pd.DataFrame({'description':filtered_sentence,"fixtype":data['Fix Type']})
```

```python
df.fixtype.unique()
```

```python
label_mapping = {'TOW Used':0,'Remote Fix':1}
df['fixtype'] = df['fixtype'].map(label_mapping)
```

```python
# df.to_csv("处理后的数据.csv")
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(df['description'].values,df['fixtype'].values,random_state =1)
```
